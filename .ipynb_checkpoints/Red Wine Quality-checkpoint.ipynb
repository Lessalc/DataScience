# Red Wine Quality

https://www.kaggle.com/uciml/red-wine-quality-cortez-et-al-2009

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
wine = pd.read_csv('https://raw.githubusercontent.com/Lessalc/DataScience/master/Data/winequality-red.csv')
```

```python
wine.columns = wine.columns.str.replace(' ', '_')
print(wine.shape)
wine.head()
```

```python
wine.describe()
```

### Primeiras Observações:
- O dataframe é formado por 11 atributos e uma classe.
- Todos os atributos são numéricos, dados quantitativos e ordinais de mensuração racional.
- Podemos ver a variação na média e no desvio padrão de cada atributo. Em Machine Learning alguns modelos usam métricas de distância para as predições, assim, essa variação pode influenciar no resultado do modelo. Isso será tratado mais a frente nesse notebook.
- Outro detalhe é que a qualidade (classe do dataframe) varia de 3 até 8. Nesse caso irei transformar de 0 até 5.

```python
classificacao = dict({3:0,4:1,5:2,6:3,7:4,8:5})
wine['quality'] = wine.quality.map(classificacao)
```

### Testando os primeiros métodos. 
- Iremos fazer um exemplo sem regularizar o dataframe, para depois comparar a evolução do modelo.
- Usaremos um pré-processamento com StandardScaler
- Um modelo de Árvore não sofre com a falta de regularização, visto que os atributos são analisados separadamente.
- Os modelos que lidam com distância e curvas que serão usados será:
    - k-NN
    - SVM
    - Regressão Logistica
- Além desses modelos, faremos teste também com os que não exigem regularizações:
    - Naive Bayes
    - Árvores de Decisão

```python
# Bibliotecas q serão usadas
# Métricas
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
# Preprocessamento
from sklearn.preprocessing import StandardScaler
# Seleção
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
# Modelos ML
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
```

```python
wine.head()
```

```python
X = wine.drop('quality', axis = 1).values
y = wine['quality'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify = wine.iloc[:,-1], random_state=0)
```

```python
# Contagem da amostra de treino e teste para as diversas qualidades
for i in range(6):
    soma_treino = 0
    soma_teste = 0
    for j in range(y_train.shape[0]):
        if y_train[j] == i:
            soma_treino += 1
    for j in range(y_test.shape[0]):
        if y_test[j] == i:
            soma_teste += 1
    print('Amostra de treino, contém: ', soma_treino, ' valores= ', i)
    print('Amostra de teste, contém: ', soma_teste, ' valores= ', i)
```

```python
print(range(y_train.shape[0]))
print(range(y_test.shape[0]))
```

### k-NN

```python
training_accuracy = []
test_accuracy = []
# Tentaremos n vizinhos num range de 2-40, pulando de dois em dois
neighbors_settings = range(2, 90, 3)
for n_neighbors in neighbors_settings:
 # construindo o modelo
 clf = KNeighborsClassifier(n_neighbors=n_neighbors)
 clf.fit(X_train, y_train)
 # Salvando acurácia de treino
 training_accuracy.append(clf.score(X_train, y_train))
 # Salvando acurácia de teste
 test_accuracy.append(clf.score(X_test, y_test))
plt.plot(neighbors_settings, training_accuracy, label="Acurácia de Treino")
plt.plot(neighbors_settings, test_accuracy, label="Acurácia de Teste")
plt.ylabel("Acurácia")
plt.xlabel("Num Vizinhos")
plt.grid()
plt.legend();
```

```python
# Escalonando os dados
# Observação, criamos uma classe chamada scale, ela irá usar os dados do treino para treinar e transformar
# Não devemos treinar e transformar nos dois dados de treino e teste, para não mudar a média e desvio que serão usados no escalonamento
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Com os dados escalonados podemos repetir o processo acima nos novos dados

training_accuracy = []
test_accuracy = []
# Tentaremos n vizinhos num range de 2-40, pulando de dois em dois
neighbors_settings = range(2, 90, 3)
for n_neighbors in neighbors_settings:
    # construindo o modelo
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train_scaled, y_train)
    # Salvando acurácia de treino
    training_accuracy.append(clf.score(X_train_scaled, y_train))
    # Salvando acurácia de teste
    test_accuracy.append(clf.score(X_test_scaled, y_test))
plt.plot(neighbors_settings, training_accuracy, label="Acurácia de Treino")
plt.plot(neighbors_settings, test_accuracy, label="Acurácia de Teste")
plt.ylabel("Acurácia")
plt.xlabel("Num Vizinhos")
plt.grid()
plt.legend();
```

```python
# Por fim, vamos ver as métricas para um n_neighbors = 80
n_neighbors = 80
# Sem escala
clf_unscaled = KNeighborsClassifier(n_neighbors=n_neighbors)
clf_unscaled.fit(X_train, y_train)
y_pred_unscaled = clf_unscaled.predict(X_test)
# Com escala
clf_scaled = KNeighborsClassifier(n_neighbors=n_neighbors)
clf_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = clf_scaled.predict(X_test_scaled)
# Métricas
print('Acurácia sem Scale: {}'.format(clf_unscaled.score(X_test, y_test)))
print('Acurácia com Scale: {}'.format(clf_scaled.score(X_test_scaled, y_test)))
print('Relatório de Classificação sem Scale')
print(classification_report(y_test, y_pred_unscaled))
print('Relatório de Classificação com Scale')
print(classification_report(y_test, y_pred_scaled))
print('Matriz de Confusão sem Scale')
print(confusion_matrix(y_test, y_pred_unscaled))
print('Matriz de Confusão com Scale')
print(confusion_matrix(y_test, y_pred_scaled))
```

```python
# Por fim, vamos ver as métricas para um n_neighbors = 80
n_neighbors = 80
# Sem escala
clf_unscaled = KNeighborsClassifier(n_neighbors=n_neighbors)
clf_unscaled.fit(X_train, y_train)
y_pred_unscaled = clf_unscaled.predict(X_test)
# Com escala
clf_scaled = KNeighborsClassifier(n_neighbors=n_neighbors)
clf_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = clf_scaled.predict(X_test_scaled)
# Métricas
print('Acurácia sem Scale: {}'.format(clf_unscaled.score(X_test, y_test)))
print('Acurácia com Scale: {}'.format(clf_scaled.score(X_test_scaled, y_test)))
print('Relatório de Classificação sem Scale')
print(classification_report(y_test, y_pred_unscaled))
print('Relatório de Classificação com Scale')
print(classification_report(y_test, y_pred_scaled))
print('Matriz de Confusão sem Scale')
print(confusion_matrix(y_test, y_pred_unscaled))
print('Matriz de Confusão com Scale')
print(confusion_matrix(y_test, y_pred_scaled))
```

### SVM
- Entre os diversos parâmetros de SVM temos alguns interessantes para testar:
    - C: Regularização dos parâmetros valor >0 e quanto menor, mais regularização
    - kernel - Tipo de algoritmo usado ('linear', 'poly', 'rbf', 'sigmoid', 'precomputed'. Por padrão 'rbf'

```python
# Primeiro Teste sem nenhum parâmetro definido
clf = SVC()
clf.fit(X_train, y_train)
print("Score de teste: {:.2f}".format(clf.score(X_test, y_test)))
```

```python
training_accuracy_modelo = []
test_accuracy_modelo = []
c_range = range(1, 100, 5)
for c in c_range:

    clf1 = SVC(C = c/10)
    clf1.fit(X_train, y_train)
    # Salvando acurácia de treino
    training_accuracy_modelo.append(clf1.score(X_train, y_train))
    # Salvando acurácia de teste
    test_accuracy_modelo.append(clf1.score(X_test, y_test))

plt.plot(c_range, training_accuracy_modelo, label="Acurácia de Treino")
plt.plot(c_range, test_accuracy_modelo, label="Acurácia de Teste")
plt.ylabel("Acurácia")
plt.xlabel("Valor de C x 10")
plt.grid()
plt.legend();
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

training_accuracy_modelo01 = []
test_accuracy_modelo01 = []

c_range = range(1, 100, 5)
for c in c_range:

    clf1 = SVC(C = c/10)
    clf1.fit(X_train, y_train)
    # Salvando acurácia de treino
    training_accuracy_modelo01.append(clf1.score(X_train_scaled, y_train))
    # Salvando acurácia de teste
    test_accuracy_modelo01.append(clf1.score(X_test_scaled, y_test))

plt.plot(c_range, training_accuracy_modelo01, label="Acurácia de Treino")
plt.plot(c_range, test_accuracy_modelo01, label="Acurácia de Teste")
plt.ylabel("Acurácia")
plt.xlabel("Valor de C x 10")
plt.grid()
plt.legend();
```

```python
training_accuracy_modelo = []
test_accuracy_modelo = []
training_accuracy_modelo01 = []
test_accuracy_modelo01 = []
lista = ['linear', 'poly', 'rbf', 'sigmoid']

for kernel in lista:
    clf = SVC(C = 10, kernel = kernel)
    clf.fit(X_train, y_train)
    # Salvando acurácia de treino
    training_accuracy_modelo.append(clf.score(X_train, y_train))
    # Salvando acurácia de teste
    test_accuracy_modelo.append(clf.score(X_test, y_test))
    
    clf1 = SVC(C = 1, kernel = kernel)
    clf1.fit(X_train_scaled, y_train)
    # Salvando acurácia de treino
    training_accuracy_modelo01.append(clf1.score(X_train_scaled, y_train))
    # Salvando acurácia de teste
    test_accuracy_modelo01.append(clf1.score(X_test_scaled, y_test))

plt.plot(lista, training_accuracy_modelo, label="Acurácia de Treino não Escalonado")
plt.plot(lista, test_accuracy_modelo, label="Acurácia de Teste não Escalonado")
plt.plot(lista, training_accuracy_modelo01, label="Acurácia de Treino Escalonado")
plt.plot(lista, test_accuracy_modelo01, label="Acurácia de Teste Escalonado")
plt.ylabel("Acurácia")
plt.xlabel("Método")
plt.grid()
plt.legend();
```

### Regressão Logística
- Iremos mais diretamente. Usaremos um modelo regularizado e um não regularizado

```python
# Sem escala
logreg_unscaled = LogisticRegression()
logreg_unscaled.fit(X_train, y_train)
y_pred_unscaled = logreg_unscaled.predict(X_test)
# Com escala
logreg_scaled = LogisticRegression()
logreg_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = logreg_scaled.predict(X_test_scaled)
# Métricas
print('Acurácia sem Scale: {}'.format(logreg_unscaled.score(X_test, y_test)))
print('Acurácia com Scale: {}'.format(logreg_scaled.score(X_test_scaled, y_test)))
print('Relatório de Classificação sem Scale')
print(classification_report(y_test, y_pred_unscaled))
print('Relatório de Classificação com Scale')
print(classification_report(y_test, y_pred_scaled))
print('Matriz de Confusão sem Scale')
print(confusion_matrix(y_test, y_pred_unscaled))
print('Matriz de Confusão com Scale')
print(confusion_matrix(y_test, y_pred_scaled))
```

### Para finalizar os métodos que usam distância, usaremos o VotingClassifier juntando os três métodos

```python
knn = KNeighborsClassifier(n_neighbors=80)
svc = SVC(C = 1, kernel = kernel)
logreg = LogisticRegression()
classificadores = [('K Nearest Neighbors',knn),('Suport Verctor Classifier',svc),('Logistic Regression',logreg)]

vc = VotingClassifier(estimators=classificadores)
vc.fit(X_train_scaled, y_train)
y_pred = vc.predict(X_test_scaled)

print("Acuracia nos dados de treino: {:.3f}".format(vc.score(X_train, y_train)))
print("Acuracia nos dados de teste: {:.3f}".format(vc.score(X_test, y_test))) 
print(confusion_matrix(y_test, y_pred))
```

## Usando modelo de Naive Bayes, Árvore de Decisão e Floresta Randomica
- Nesse modelo a normatização não é necessária.

```python
# Gerando o modelo classificador de Naive Bayes
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Acuracia nos dados de treino: {:.3f}".format(clf.score(X_train, y_train)))
print("Acuracia nos dados de teste: {:.3f}".format(clf.score(X_test, y_test))) 
print(confusion_matrix(y_test, y_pred))
```

```python
# Usando Árvore de Decisão, primeiramente com o processo livre, sem impor limitações, o que pode gerar modelos super ajustados
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
print("Acuracia nos dados de treino: {:.3f}".format(tree.score(X_train, y_train)))
print("Acuracia nos dados de teste: {:.3f}".format(tree.score(X_test, y_test))) 
print(confusion_matrix(y_test, y_pred))
```

```python
# O modelo livre resultou exatamente em um modelo super ajustado. Vamos ajustar os parâmetros
# Quantidade mínima de amostras nas folhas
min_leaf = [5, 10, 20, 30, 50] 
# Profundidade da árvore
max_prof = [4,6,8,10] # (4,6,8,10) 

training_accuracy_modelo = []
test_accuracy_modelo = []
qtd_leaf = []
profundidade = []


for folha in min_leaf:
    qtd_leaf.append(folha)
    for camada in max_prof:
        profundidade.append(camada)
        tree = DecisionTreeClassifier(max_depth= camada, min_samples_leaf = folha, random_state=0)
        tree.fit(X_train, y_train)
        y_pred = tree.predict(X_test)
        training_accuracy_modelo.append(tree.score(X_train, y_train))
        test_accuracy_modelo.append(tree.score(X_test, y_test))
    
```

```python
camadas4_treino, camadas6_treino, camadas8_treino, camadas10_treino = [],[],[],[]
camadas4_teste, camadas6_teste, camadas8_teste, camadas10_teste = [],[],[],[]
for i, j in enumerate(profundidade):
    if j == 4:
        camadas4_treino.append(training_accuracy_modelo[i])
        camadas4_teste.append(test_accuracy_modelo[i])
    elif j == 6:
        camadas6_treino.append(training_accuracy_modelo[i])
        camadas6_teste.append(test_accuracy_modelo[i])
    elif j == 8:
        camadas8_treino.append(training_accuracy_modelo[i])
        camadas8_teste.append(test_accuracy_modelo[i])
    elif j == 10:
        camadas10_treino.append(training_accuracy_modelo[i])
        camadas10_teste.append(test_accuracy_modelo[i])
```

```python
plt.figure(figsize=(15, 6))

plt.plot(qtd_leaf, camadas4_treino, ls='--' ,label="Acurácia de Treino 4 camadas de Profundidade")
plt.plot(qtd_leaf, camadas4_teste, label="Acurácia de Teste 4 camadas de Profundidade")

plt.plot(qtd_leaf, camadas6_treino, ls='--' ,label="Acurácia de Treino 6 camadas de Profundidade")
plt.plot(qtd_leaf, camadas6_teste, label="Acurácia de Teste 6 camadas de Profundidade")

plt.plot(qtd_leaf, camadas8_treino, ls='--' ,label="Acurácia de Treino 8 camadas de Profundidade")
plt.plot(qtd_leaf, camadas8_teste, label="Acurácia de Teste 8 camadas de Profundidade")

plt.plot(qtd_leaf, camadas10_treino, ls='--' ,label="Acurácia de Treino 10 camadas de Profundidade")
plt.plot(qtd_leaf, camadas10_teste, label="Acurácia de Teste 10 camadas de Profundidade")


plt.ylabel("Acurácia")
plt.xlabel("Quantidade mínima de amostras nas folhas")
plt.grid()
plt.legend();
```

```python
# Escolhendo um modelo, dessa vez sem limitar as amostras
tree = DecisionTreeClassifier(max_depth= 8, random_state=2)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
print("Acuracia nos dados de treino: {:.3f}".format(tree.score(X_train, y_train)))
print("Acuracia nos dados de teste: {:.3f}".format(tree.score(X_test, y_test)))
print(confusion_matrix(y_test, y_pred))
```

#### Floresta Randomica, usando mesmo modelo final

```python
# Limitado um minímo de amostras nas folhas
forest = RandomForestClassifier(n_estimators=200, max_depth = 8, min_samples_leaf = 0.05, random_state=2)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
print("Acuracia nos dados de treino: {:.3f}".format(forest.score(X_train, y_train)))
print("Acuracia nos dados de teste: {:.3f}".format(forest.score(X_test, y_test)))
print(confusion_matrix(y_test, y_pred))
```

```python
# sem limitar amostas nas folhas
forest = RandomForestClassifier(n_estimators=200, max_depth = 8, random_state=2)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
print("Acuracia nos dados de treino: {:.3f}".format(forest.score(X_train, y_train)))
print("Acuracia nos dados de teste: {:.3f}".format(forest.score(X_test, y_test)))
print(confusion_matrix(y_test, y_pred))
```

```python
def plot_importancia_atributos(modelo, titulo="Título do Gráfico"):
    atributos = wine.iloc[:,:-1].columns.shape[0]
    plt.barh(range(atributos), modelo.feature_importances_, align='center')
    plt.yticks(np.arange(atributos), wine.iloc[:,:-1].columns)
    plt.title(titulo)
    plt.xlabel("Importância dos Atributos")
    plt.ylabel("Atributos")
```

```python
plot_importancia_atributos(forest, "Modelo - Floresta")
```

### Conclusões
- Nosso conjunto é grande em quantidade de dados, no entanto, algumas das classes possuem uma quantidade muito pequena de dados, o que dificultou os modelos de identificá-las.
- A acurácia nos dá uma bo ideia da qualidade do modelo, no entanto, para classificações é visivelmente melhor e mais importante observar a matriz de confusão ou o relatório de classificação.
- No nosso caso particular é uma simles classificação de nota para vinhos, a matriz gera uma observação interessante, mas caso fosse uma identificação de doença ou risco, por exemplo, a matriz torna-se imprescindível, visto que identificar falsos positivos e negativos é crucial nessa análise, ou identificar onde estão os erros (caso fosse uma classificação de risco).
